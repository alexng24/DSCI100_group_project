The ATP Tour comprises ATP Masters 1000, ATP 500, and ATP 250, and the higher the numbers behind, the higher rank of the tournaments. We’ve done research on the probability of winning, and it is mentioned that playing the first serve successfully helps to win the game. Hence, our project will use the ATP tour data from 2017 to 2019 to analyze whether the players’ age, height and performance during the game is related to the result of the game. 
We would like to predict whether or not a player can win the game based on their height, age and performance during the game. 
We will construct several models using different predictors, and include the best  two predictors into our final model.
The question we want to answer is if we can build a classifier to accurately predict the outcome of a tennis match based on height, age, and other tennis statistics.

```r
# Import Libaries
library(tidyverse)
library(tidymodels)
library(repr)
library(rvest)
library(stringr)
options(repr.matrix.max.rows = 6)
```

```r
# Load and mutate data
# tennis_data <- read_csv('atp2017-2019.csv')
tennis_data_2019<-read.csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2019.csv')
tennis_data_2018<-read.csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2018.csv')
tennis_data_2017<-read.csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2019.csv')
tennis_data <- bind_rows(tennis_data_2019, tennis_data_2018, tennis_data_2017)
winner_data <- select(tennis_data, c(winner_name, 
                      winner_age, 
                      winner_ht, 
                      w_ace, 
                      w_df, 
                      w_svpt, 
                      w_1stIn, 
                      w_1stWon, 
                      w_SvGms, 
                      w_bpSaved, 
                      w_bpFaced)) %>%
                rename(name=winner_name,
                      age=winner_age,
                      height=winner_ht,
                      aces=w_ace,
                      double_faults=w_df,
                      serve_points=w_svpt,
                      first_serves_made=w_1stIn,
                      first_serve_points=w_1stWon,
                      serve_games=w_SvGms,
                      break_points_saved=w_bpSaved,
                      break_points_faced=w_bpFaced) %>%
                add_column(match_result='win')
loser_data <- select(tennis_data, c(loser_name, 
                      loser_age, 
                      loser_ht, 
                      l_ace, 
                      l_df, 
                      l_svpt, 
                      l_1stIn, 
                      l_1stWon, 
                      l_SvGms, 
                      l_bpSaved, 
                      l_bpFaced)) %>%
                rename(name=loser_name,
                      age=loser_age,
                      height=loser_ht,
                      aces=l_ace,
                      double_faults=l_df,
                      serve_points=l_svpt,
                      first_serves_made=l_1stIn,
                      first_serve_points=l_1stWon,
                      serve_games=l_SvGms,
                      break_points_saved=l_bpSaved,
                      break_points_faced=l_bpFaced) %>%
                add_column(match_result='lose')
tennis_data_mod <- bind_rows(winner_data, loser_data)
```

```r
# Split Data into Train and Test Sets

tennis_split <- initial_split(tennis_data_mod, prop=0.75, strata=match_result)
tennis_train <- training(tennis_split)
tennis_test <- testing(tennis_split)
```

```r
# mean for each predictor
predictors_mean<- map_dfr(tennis_train[,2:10], mean, na.rm=TRUE)
predictors_mean

# distribution of match results in training set
match_result_dist <- tibble(num_of_wins=nrow(filter(tennis_train, match_result=='win')), num_of_loses=nrow(filter(tennis_train, match_result=='lose')))
match_result_dist

# count of NA's in each predictor column
count_of_nas <- map_dfr(tennis_train[,2:10], ~sum(is.na(.)))
count_of_nas
```

```r
tennis_train_plot_1 <- tennis_train %>%
                            ggplot(aes(x=age, y=height,color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Height (cm)", color='Match Result')

tennis_train_plot_2 <- tennis_train %>%
                            ggplot(aes(x=age, y=aces, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of Aces", color='Match Result')

tennis_train_plot_3 <- tennis_train %>%
                            ggplot(aes(x=age, y=double_faults, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of Double Faults", color='Match Result')

tennis_train_plot_4 <- tennis_train %>%
                            ggplot(aes(x=age, y=serve_points, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of Serve Points", color='Match Result')

tennis_train_plot_5 <- tennis_train %>%
                            ggplot(aes(x=age, y=first_serves_made, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of First Serves Made", color='Match Result')

tennis_train_plot_6 <- tennis_train %>%
                            ggplot(aes(x=age, y=first_serve_points, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of First Serve Points", color='Match Result')

tennis_train_plot_7 <- tennis_train %>%
                            ggplot(aes(x=age, y=serve_games, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of Serve Games", color='Match Result')

tennis_train_plot_8 <- tennis_train %>%
                            ggplot(aes(x=age, y=break_points_saved, color=match_result)) +
                            geom_point(alpha=0.2) +
                            labs(x="Age (years)", y="Number of Break Points Saved", color='Match Result')


tennis_train_plot_1
tennis_train_plot_2
tennis_train_plot_3
tennis_train_plot_4
tennis_train_plot_5
tennis_train_plot_6
tennis_train_plot_7
tennis_train_plot_8
```

## Methods

The data analysis will be conducted using a K-nearest neighbors model, with only certain columns from the data set being included in the analysis. The columns that will be used initially are a player’s age, height and several statistics specific to a winner’s match performance: w_ace, w_df, w_svpt, w_1stIn, w_1stWon, w_2ndWon, w_SvGms, w_bpSaved and w_bpFaced. From these variables, forward selection will be performed to create the KNN model with up to 3 predictors that provides the greatest prediction accuracy. Cross-validation will then be performed to determine the number of neighbors (K) that maximizes accuracy. The results will then be visualized on a scatterplot with the predictor variables on the axes and the points colour-coded for match winners.

## Expected Outcomes and Significance


Through this analysis, we expect to find players with higher positive tennis stats (ex. higher number of aces, service points), greater height, and younger age would be more likely to win and vice versa.
Our findings could help discover which factors are most important for winning in professional tennis matches. Based on the importance of the tennis stats, coaches could help build an efficient training plan for athletes.
